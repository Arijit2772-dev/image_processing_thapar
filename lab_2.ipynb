```python
img = cv2.imread("/content/demo.jpg",cv2.IMREAD_GRAYSCALE)
```

```python
cv2_imshow(img)
```

```python
# convert image to float for computation
img_float = img.astype(np.float32)
```

```python
# computing scaling constant c
c = 255/np.log(1+np.max(img_float))
```

```python
# apply log transformation
log_img = c*np.log1p(img_float) #log(1+r)
```

```python
# convert back to 8-bit image
log_img = np.uint8(log_img)
```

```python
cv2_imshow(log_img)
```
